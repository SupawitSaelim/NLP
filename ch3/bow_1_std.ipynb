**Simple Topic Identification**

Word counts with bag-of-words
*   Bag of words is a basic method to finding topics in a text.
*   More frequent a word, it might be the more important in a text.




```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
from collections import Counter
text = """  The cat is in the box.
            The cat likes the box.
            The box is over the cat."""
counter = Counter(word_tokenize(text))
print(counter)
count = counter.most_common(2)
print(count)
```

**Simple text preprocessing**
: Text processing helps make for better input data when performing machine learning or other statistical methods.

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
text = """The cat is in the box. The cat likes the box. The box is over the cat in the house."""
tokens = [w for w in word_tokenize(text.lower()) if w.isalpha()]
no_stops = [t for t in tokens if t not in stopwords.words('english')]
print(Counter(no_stops))
```
